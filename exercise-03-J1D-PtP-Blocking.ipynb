## Description

In this exercise we consider the Laplace equation in one dimension,
namely $$\frac{d^2V}{dx^2}=0$$
with Dirichlet boundary conditions, $V_{x=0}=0$ and $V_{x=1}=1$. It has a very simple analytical solution, $V(x)=x$. Here we use the <font color='green'>Jacobi</font> method to solve it numerically so that we have the following <font color='green'>stencil</font> loop over all interior grid points:

$$\forall i: V_i=\frac{V_{i+1}+V_{i-1}}{2}$$

By repeating the loop for many iterations, the numerical solution will converge, though in this exercise we do not seek how efficient and accurate the method is. Instead, we investigate how <font color='green'>domain decomposition</font> can be used to distribute the workload across all MPI processes by splitting the array of the potential function (chunk of grid points). The ghost cell exchange is applied using MPI point-to-point communication. Indeed, halo exchange is trivial in this exercise since there is only one element (scalar) on each side of the chunk.

- In the directory `J1D-PtP-Blocking`, there are **f** and **c** subdirectories for Fortran and C, respectively. If you have loaded the modules as described in exercise **Hello**, compilers and MPI wrappers are ready for use. The compilation process is facilitated using a `Makefile`.
- `MPI_PROC_NULL` is a constant that represents a dummy MPI process rank. When passing it to a <font color='green'>send</font> or <font color='green'>receive</font> operation for instance, it is guaranteed to succeed and return as soon as possible with noting sent and no modifications to the receive buffer.
- The workload distribution is done by the  function/subroutine `get_share_per_proc`, which determines the size of the chunk (and in the case of Fortran also the index of start and end grid points).
