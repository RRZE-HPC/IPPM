## Description

- Most exercises provide some code to start with. These codes are distributed across several folders within the folder `code` and are should be already in your home directory, assuming you cloned the **IPPM** from GitHuB.



- In this exercise we use the code in folder **Hello**. We want to initialize MPI, print out `size`/`rank`, and finalize the MPI. There are **7** `FIXME` markers in the C code (**5** in the case of the Fortran code) which are related to the name of calls to MPI initialization and finalization as well as obtaining the size of a communicator. Some `FIXME` markers concern arguments of the calls. Your task is to replace the `FIXME` markers with the correct code.

- In the directory `Hello`, there are `f` and `c` subdirectories for Fortran and C, respectively.

- To make Intel MPI and compilers available, you should load two modules in the <font color='green'>Terminal</font>:
```bash
module load intel
module load intelmpi
```
- In order to compile an MPI program, you have to use one of the MPI wrapper scripts of the MPI library instead of directly running the compilers, e.g. in the case of Intel MPI, `mpiicc`, `mpiicpc`, and `mpiifort` for C, C++, and Fortran codes, respectively. In this case exercise:
  - For C++:
    ```bash
    mpiicpc hello.cpp
    ```
  - For Fortran:
    ```bash
    mpiifort hello.F90
    ```
- For running the code you should use mpirun: `mpirun -np # ./my_executable` <br>
  - If the compilation and building the binary is done as explained above, the name of the executable will be `a.out`.
  - Here, `#` is the number of processes you want to use. Each Fritz compute node has **72** cores. You can try running with different number of MPI processes up to **72**.
